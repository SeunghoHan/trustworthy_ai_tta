## 1. Decision-based Boundary Attack

**API Reference:**
- https://adversarial-robustness-toolbox.readthedocs.io/en/latest/modules/attacks/evasion.html#boundary-attack-decision-based-attack

**Related paper:**
- https://arxiv.org/abs/1712.04248

---

```python
import tensorflow as tf
tf.compat.v1.disable_eager_execution() # To use the ART

from tensorflow.keras import datasets, layers, models

from art.utils import load_dataset
from art.estimators.classification import KerasClassifier

import matplotlib.pyplot as plt
import numpy as np
import os
os.environ["CUDA_VISIBLE_DEVICES"]="0"

cifar_model_path = '/archive/workspace/TTA/demo/models/baisc_cnn_cifar'
```

### Load dadaset (CIFAR10)

```python
(x_train, y_train), (x_test, y_test), min_, max_ = load_dataset(str("cifar10"))
x_val, y_val = x_train[40000:], y_train[40000:]
x_train, y_train = x_train[:40000], y_train[:40000]

class_name = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'ship', 'truck']
```

```python
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i], cmap=plt.cm.binary)
```

### Load pre-trained model

```python
trained_model = models.load_model(cifar_model_path)
classifier = KerasClassifier(model=trained_model, clip_values=(0, 1))
```

```python
loss_test, accuracy_test = trained_model.evaluate(x_test, y_test)
print('Accuracy on test data(Before attack): {:4.2f}%'.format(accuracy_test * 100))
```

### Generate adversairal examples

```python
from art.attacks.evasion import FastGradientMethod

# eps: Attack step size (input variation)
attack_fgsm = FastGradientMethod(estimator=classifier, eps=0.3)
x_test_adv = attack_fgsm.generate(x_test)
```

```python
loss_test, accuracy_test = trained_model.evaluate(x_test_adv, y_test)
perturbation = np.mean(np.abs((x_test_adv - x_test)))
print('Accuracy on adversarial test data: {:4.2f}%'.format(accuracy_test * 100))
print('Average perturbation: {:4.2f}'.format(perturbation))
```

### Generate adversairal examples

```python
from art.attacks.evasion import BoundaryAttack

attack_ba_unt = BoundaryAttack(
    estimator=classifier, 
    targeted=False, 
    max_iter=0, 
    delta=0.001, 
    epsilon=0.001
)

x_test_adv = attack_ba_unt.generate(x_test[:20])
```

```python
loss_test, accuracy_test = trained_model.evaluate(x_test_adv, y_test[:20])
perturbation = np.mean(np.abs((x_test_adv - x_test[:20])))
print('Accuracy on adversarial test data: {:4.2f}%'.format(accuracy_test * 100))
print('Average perturbation: {:4.2f}'.format(perturbation))
```

```python
target_img = 16

preds_origin = trained_model.predict(x_test)
preds_adv = trained_model.predict(x_test_adv)

plt.subplot(1, 2, 1)
plt.imshow(x_test[target_img])
plt.title('original img')
      
plt.subplot(1, 2, 2)
plt.imshow(x_test_adv[target_img])
plt.title('adversarial img')

print('Prediction for original img: {}'.format(class_name[np.argmax(preds_origin[target_img])]))
print('Prediction for adversairal img: {}'.format(class_name[np.argmax(preds_adv[target_img])]))
```

## 3. Zeroth Order Optimisation (ZOO)

**API Reference:**
- https://adversarial-robustness-toolbox.readthedocs.io/en/latest/modules/attacks/evasion.html#zeroth-order-optimization-zoo-attack

**Related paper:**
- https://arxiv.org/abs/1708.03999

For the Catboost model, train the model with MNIST dataset and perform the ZOO attack to the trained model. 

---  

### Load dataset (MNIST)

```python
from art.utils import load_mnist

(x_train_mn, y_train_mn), (x_test_mn, y_test_mn), min_, max_ = load_mnist()

n_samples_train = x_train_mn.shape[0]
n_features_train = x_train_mn.shape[1] * x_train_mn.shape[2] * x_train_mn.shape[3]
n_samples_test = x_test_mn.shape[0]
n_features_test = x_test_mn.shape[1] * x_test_mn.shape[2] * x_test_mn.shape[3]

x_train_mn = x_train_mn.reshape(n_samples_train, n_features_train)
x_test_mn = x_test_mn.reshape(n_samples_test, n_features_test)

y_train_mn = np.argmax(y_train_mn, axis=1)
y_test_mn = np.argmax(y_test_mn, axis=1)

n_samples_max = 5000
x_train_mn = x_train_mn[0:n_samples_max]
y_train_mn = y_train_mn[0:n_samples_max]
x_test_mn = x_test_mn[0:n_samples_max]
y_test_mn = y_test_mn[0:n_samples_max]
```

### Train model with CatBoostClassifier

```python
from catboost import CatBoostClassifier

model = CatBoostClassifier(custom_loss=['Accuracy'], random_seed=42, logging_level='Silent', depth=8, iterations=20)
model.fit(x_train_mn, y_train_mn, cat_features=None, eval_set=(x_train_mn, y_train_mn))
```

```python
y_pred = model.predict_proba(x_test_mn)
score = np.sum(y_test_mn == np.argmax(y_pred, axis=1)) / y_test_mn.shape[0]
print("Benign Training Score: %.4f" % score)
```

### Generate adversarial examples

```python
from art.estimators.classification import CatBoostARTClassifier

cb_classifier = CatBoostARTClassifier(model=model, nb_features=x_train_mn.shape[1])
```

```python
from art.attacks.evasion import ZooAttack

attack_zoo = ZooAttack(
    classifier=cb_classifier, 
    confidence=0.0, 
    targeted=False, 
    learning_rate=1e-1, 
    max_iter=50,
    binary_search_steps=20, 
    initial_const=1e-3, 
    abort_early=True, 
    use_resize=False, 
    use_importance=False, 
    nb_parallel=10, 
    batch_size=1, 
    variable_h=0.25
)
```

```python
x_test_adv = attack_zoo.generate(x_test_mn[:20])
```

```python
y_pred = model.predict_proba(x_test_adv)
score = np.sum(y_test_mn[:20] == np.argmax(y_pred, axis=1)) / y_test_mn[:20].shape[0]
print("Benign Training Score: %.4f" % score)
```

```python
preds_origin = np.argmax(model.predict_proba(x_test_mn[0:1, :]), axis=1)
preds_adv = np.argmax(model.predict_proba(x_test_adv[0:1, :]), axis=1)

plt.subplot(1, 2, 1)
plt.imshow(x_test_mn[0, :].reshape((28, 28)))
plt.title('origianl image')

plt.subplot(1, 2, 2)
plt.imshow(x_test_adv[0, :].reshape((28, 28)))
plt.title('adversarial image')

print('Prediction for original img: {}'.format(preds_origin))
print('Prediction for adversairal img: {}'.format(preds_adv))
```
