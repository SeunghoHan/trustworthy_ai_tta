## 1. Fast Gradient Sign Method(FGSM) 

**API Reference:**
- https://adversarial-robustness-toolbox.readthedocs.io/en/latest/modules/attacks/evasion.html#fast-gradient-method-fgm

**Related paper:**
- https://arxiv.org/abs/1412.6572

**To use the ART FGSM attack, follow the procedure below.**

1. Load the trained model and create ART Keras classifier using it.
1. Create ART FGSM attack instance and generate adversarial test data.
1. Evaluate the accuracy with adversarial test data.

---

```python
import tensorflow as tf
tf.compat.v1.disable_eager_execution() # To use the ART

from tensorflow.keras import datasets, layers, models

from art.utils import load_dataset
from art.estimators.classification import KerasClassifier

import matplotlib.pyplot as plt
import numpy as np
import os

cifar_model_path = '/archive/workspace/TTA/demo/models/baisc_cnn_cifar'
```

### Load dadaset (CIFAR10)

```python
(x_train, y_train), (x_test, y_test), min_, max_ = load_dataset(str("cifar10"))
x_val, y_val = x_train[40000:], y_train[40000:]
x_train, y_train = x_train[:40000], y_train[:40000]

class_name = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'ship', 'truck']
```

```python
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i], cmap=plt.cm.binary)
```

### Load pre-trained model

```python
trained_model = models.load_model(cifar_model_path)
classifier = KerasClassifier(model=trained_model, clip_values=(0, 1))
```

```python
loss_test, accuracy_test = trained_model.evaluate(x_test, y_test)
print('Accuracy on test data(Before attack): {:4.2f}%'.format(accuracy_test * 100))
```

### Generate adversairal examples

```python
from art.attacks.evasion import FastGradientMethod

# eps: Attack step size (input variation)
attack_fgsm = FastGradientMethod(estimator=classifier, eps=0.3)
x_test_adv = attack_fgsm.generate(x_test)
```

```python
loss_test, accuracy_test = trained_model.evaluate(x_test_adv, y_test)
perturbation = np.mean(np.abs((x_test_adv - x_test)))
print('Accuracy on adversarial test data: {:4.2f}%'.format(accuracy_test * 100))
print('Average perturbation: {:4.2f}'.format(perturbation))
```

```python
img_num = 10

preds_origin = trained_model.predict(x_test)
preds_adv = trained_model.predict(x_test_adv)

plt.subplot(1, 2, 1)
plt.imshow(x_test[img_num])
plt.title('original img')
      
plt.subplot(1, 2, 2)
plt.imshow(x_test_adv[img_num])
plt.title('adversarial img')


print('Prediction for original img: {}'.format(class_name[np.argmax(preds_origin[img_num])]))
print('Prediction for adversairal img: {}'.format(class_name[np.argmax(preds_adv[img_num])]))
```

## 4. Feature Adversaries

**API Reference:**
- https://adversarial-robustness-toolbox.readthedocs.io/en/latest/modules/attacks/evasion.html#projected-gradient-descent-pgd-tensorflowv2

**Related paper:**
- https://arxiv.org/abs/1706.06083

--- 

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPool2D

from art.utils import load_mnist
from art.estimators.classification import TensorFlowV2Classifier

import matplotlib.pyplot as plt
import numpy as np
import os

mnist_model_path = '/archive/workspace/TTA/demo/models/baisc_cnn_mnist'
```

### Load dadaset (MNIST)

```python
(x_train, y_train), (x_test, y_test), min_pixel_value, max_pixel_value = load_mnist()

# Step 1a: Cast to np.float32
x_train = x_train.astype(np.float32)
x_test = x_test.astype(np.float32)
```

### Train model

```python
model = Sequential()
model.add(Conv2D(filters=4, kernel_size=5, activation="relu"))
model.add(MaxPool2D(pool_size=(2, 2), padding="valid", data_format=None))
model.add(Conv2D(filters=10, kernel_size=5, activation="relu"))
model.add(MaxPool2D(pool_size=(2, 2), padding="valid", data_format=None))
model.add(Flatten())
model.add(Dense(100))
model.add(Dense(10))
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
loss_object = tf.keras.losses.CategoricalCrossentropy(from_logits=True)

def train_step(model, images, labels):
    with tf.GradientTape() as tape:
        predictions = model(images, training=True)
        loss = loss_object(labels, predictions)
    gradients = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))
```

```python
classifier = TensorFlowV2Classifier(
    model=model,
    loss_object=loss_object,
    train_step=train_step,
    nb_classes=10,
    input_shape=(28, 28, 1),
    clip_values=(0, 1),
)


classifier.fit(x_train, y_train, batch_size=64, nb_epochs=3)
```

```python
predictions = classifier.predict(x_test)
accuracy = np.sum(np.argmax(predictions, axis=1) == np.argmax(y_test, axis=1)) / len(y_test)
print("Accuracy on benign test examples: {}%".format(accuracy * 100))
```

### Generate adversairal examples

```python
from art.attacks.evasion import FeatureAdversariesTensorFlowV2

# Step 5: prepare a batch of source and guide images
valid = np.argmax(y_test, axis=1)[:100] != np.argmax(y_test, axis=1)[100:200]
source = x_test[:100][valid][:32]
guide = x_test[100:200][valid][:32]
```

```python
attack_fa = FeatureAdversariesTensorFlowV2(
    classifier,
    layer=-2,
    delta=45/255,
    optimizer=None,
    step_size=1/255,
    max_iter=100,
)
x_test_adv = attack_fa.generate(source, guide)
```

```python
predictions = classifier.predict(x_test_adv)
accuracy = np.sum(np.argmax(predictions, axis=1) == np.argmax(y_test[:100][valid][:32], axis=1)) / len(y_test[:100][valid][:32])

dim = tuple(range(1, len(source.shape)))
pert = np.mean(np.amax(np.abs(source - x_test_adv), axis=dim))
print("Accuracy on adversarial test batch: {}%".format(accuracy * 100))
print("Average perturbation: {}%".format(pert))
```

```python
target_img = 2

plt.subplot(1, 2, 1)
plt.imshow(x_test[target_img])
plt.title('original img')
      
plt.subplot(1, 2, 2)
plt.imshow(x_test_adv[target_img])
plt.title('adversarial img')
```
